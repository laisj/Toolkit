```python
import matplotlib.pyplot as plt
from numpy.random import randn

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (14.0, 8.0)

fig,ax = plt.subplots(2,3)
plt.plot(randn(50).cumsum(), 'm--',label='1')
ax[0,1].plot(randn(50).cumsum(), 'k--',label='2')
ax[1,2].set_xlabel('ddddx')
ax[1,2].set_ylabel('ddddy')
ax[0,2].hist(randn(100),bins=20,color='k', alpha=0.1,label='3')
ax[1,2].scatter(np.arange(30), np.arange(30)+3*randn(30),label='4')

```
