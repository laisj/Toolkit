# 广告CTR的置信度计算
 
## CTR实验的假设与抽象

问题：广告主新建一个广告a，在一段时间的数据积累后收集到了d个曝光，c个点击，在此期间内广告a的定向与出价没有变化，并且任意两个曝光事件没有关联并且任意两次曝光情况相同（独立同分布），如何估计广告a的点击率，并判断该估计的置信度是否充足。

## 问题抽象

显然，在没有其他先验知识的情况下，对a的点击率的合理估计值为：$c/d$。但是当c与d的数量较少时，数据量不充分，此估计是不可信的。

从概率角度看，d次曝光可以看作d次实验，每次实验有p的概率获得点击，最终d次实验获得c个点击。整个实验的过程可以抽象为伯努利实验，所对应的分布为二项分布；而p的所有可能取值是“二项分布之上的分布”，为二项分布的共轭分布：Beta分布。

进一步解释：假设广告主进行多次尝试，每次的d与c值均不相同，则$c/d$的值也满足一定规律的分布，这个分布就是Beta分布。

## 二项分布与Beta分布的直观认识

以下代码为单次实验概率为0.006，实验次数为50000的二项分布的概率密度图像（图1）：

```python
# encoding=utf8  
import sys  
stdout = sys.stdout
reload(sys)  
sys.setdefaultencoding('utf8')
sys.stdout = stdout

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号

%matplotlib inline

# 定义二项分布的基本信息
p = 0.006 # 事件A概率0.4
n = 50000   # 重复实验50000次

# 伪造符合二项分布的随机变量 (random variates)
X = stats.binom.rvs(n,p,size=50000)

# 作出上面满足二项分布随机变量的频率分布直方图
plt.hist(X, 500, normed=True)
plt.title("figure 1")
plt.show()
```

以下代码为实验次数为50000，正样本为300的Beta分布的概率密度图像（图2）：

```python
X = stats.beta.rvs(301, 49701, size=50000)
 
plt.hist(X, 500, normed=True)
plt.title("figure 2")
plt.show()
```

以下代码为实验次数为500，正样本为3的Beta分布的概率密度图像（图3）：

```python
X = stats.beta.rvs(4, 498, size=50000)
 
plt.hist(X, 500, normed=True)
plt.title("figure 3")
plt.show()
```

## 点击率置信度的计算

如图2与图3所示，当点击数为300，曝光数为50000时，点击率的“范围”大致在0.005与0.007之间；而当点击数为3，曝光数为500时，点击率的“范围”在0与0.025之间。显然后者的范围“较宽”，置信度比前者差。这也是符合直觉的。

### 置信度的量化

在概率论中，通常认为发生概率小于5%的事件称为小概率事件，小概率事件的发生，通常意味着异常情况。对应于CTR探索实验中，每一次探索实验对应一个Beta分布，我们可以查看这个分布的95%置信区间，使用这个区间的宽窄来定义置信度。

一次CTR探索实验对应的Beta分布为：$Beta(click+1, diplay-click+1)$, 概率密度图与累积概率密度图如下：

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 5))
x = np.linspace(stats.beta.ppf(0.0, 4, 498), stats.beta.ppf(0.999, 4, 498), 1000)
rv = stats.beta(4, 498)
ax[0].plot(x, rv.pdf(x), 'k-', lw=1.5)
ax[0].plot([0.006, 0.006,], [0, rv.pdf(0.006),], 'k--', linewidth=2.5)
ax[0].annotate('max likelyhood: 0.006', xy=(0.006, rv.pdf(0.006)), xytext=(0.010, 110),arrowprops=dict(facecolor='blue', shrink=0.05))
plt.title("figure 4")
ax[1].plot(x, rv.cdf(x), 'k-', lw=1.5)
plt.show()
```

累积概率密度函数的反函数如下：

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 5))
x = np.linspace(stats.beta.ppf(0.0, 4, 498), stats.beta.ppf(0.999, 4, 498), 100)
y = np.linspace(0, 0.999, 1000)
rv = stats.beta(4, 498)
markers_on = [25, 975]

plt.title("figure 5")
ax[0].plot(x, rv.cdf(x), 'k-', lw=1.5)
ax[1].plot(y, rv.ppf(y), 'k-', lw=1.5)
ax[1].annotate('lower bound:' + str(rv.ppf(0.025)), xy=(0.025, rv.ppf(0.025)), xytext=(0.03, 0.01),arrowprops=dict(facecolor='blue', shrink=0.05))
ax[1].annotate('upper bound:' + str(rv.ppf(0.975)), xy=(0.975, rv.ppf(0.975)), xytext=(0.9, 0.025),arrowprops=dict(facecolor='blue', shrink=0.05))
ax[1].plot([0.025, 0.025,], [0, rv.ppf(0.025),], 'k--', linewidth=2.5)
ax[1].plot([0.975, 0.975,], [0, rv.ppf(0.975),], 'k--', linewidth=2.5)
ax[1].plot([0, 0.025,], [rv.ppf(0.025), rv.ppf(0.025),], 'k--', linewidth=2.5)
ax[1].plot([0, 0.975,], [rv.ppf(0.975), rv.ppf(0.975),], 'k--', linewidth=2.5)
plt.show()
```

可见本次CTR探索实验的上下界为（0.002，0.017）。查看点击300，曝光50000的CTR实验的置信区间上下界：

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 5))
x = np.linspace(stats.beta.ppf(0.0, 4, 498), stats.beta.ppf(0.999, 4, 498), 100)
y = np.linspace(0, 0.999, 1000)
rv = stats.beta(301, 49701)

plt.title("figure 6")
ax[0].plot(x, rv.cdf(x), 'k-', lw=1.5)
ax[1].plot(y, rv.ppf(y), 'k-', lw=1.5)
ax[1].annotate('lower bound:' + str(rv.ppf(0.025)), xy=(0.025, rv.ppf(0.025)), xytext=(0.03, 0.007),arrowprops=dict(facecolor='blue', shrink=0.05))
ax[1].annotate('upper bound:' + str(rv.ppf(0.975)), xy=(0.975, rv.ppf(0.975)), xytext=(0.9, 0.007),arrowprops=dict(facecolor='blue', shrink=0.05))
ax[1].plot([0.025, 0.025,], [0, rv.ppf(0.025),], 'k--', linewidth=2.5)
ax[1].plot([0.975, 0.975,], [0, rv.ppf(0.975),], 'k--', linewidth=2.5)
ax[1].plot([0, 0.025,], [rv.ppf(0.025), rv.ppf(0.025),], 'k--', linewidth=2.5)
ax[1].plot([0, 0.975,], [rv.ppf(0.975), rv.ppf(0.975),], 'k--', linewidth=2.5)
plt.show()
```

可见置信区间变窄了很多：（0.0053，0.0067）。

### 置信区间的计算方法

由以上例子，归纳出一般的CTR置信区间计算方法：

1. 统计CTR探索实验中的曝光数d与点击数c；
2. 得出CTR的分布：$Beta(c+1,d-c+1)$
3. 计算Beta分布的累积概率函数
4. 累积概率函数值为0.975与0.025的两个值即为置信区间的上下界。

## 满足广告主的需求

在实际环境中，广告主新建广告后，如果平台对新CTR广告的探索实验分配了过多的资源，会使得平台的利益受损；如果分配的资源不足，则可能埋没一些广告，对平台的生态也不利。所以我们需要计算，对于新广告，平台应该分配多少资源。

置信度的衡量，使用指标：$(置信区间上界-置信区间下界)/统计点击率$，此值越小，说明CTR置信度越高。

曝光50000，点击300，置信度为：$(0.0067-0.0053)/0.006 = 0.23 = 23％$，意味着实际CTR可能比观测值上下浮动约10%。

曝光500，点击3，置信度为: $(0.017-0.002)/0.006 = 2.5 = 250％$，意味着实际CTR可能比观测值上下浮动约100%。

使用网格搜索，分别查看：

1. 探索实验点击数在（10，2000）范围内，点击率在（0.001，0.05）范围内的置信区间上下界情况；
2. 探索实验曝光数在（1000，20000）范围内，点击率在（0.001，0.05）范围内的置信区间上下界情况。

```python
from mpl_toolkits.mplot3d import axes3d, Axes3D 
fig = plt.figure()
ax = Axes3D(fig)

n = np.arange(10, 2000, 10)
p = np.arange(0.001, 0.05, 0.001)
n, p = np.meshgrid(n, p)

a = n / p
b = n
zz = stats.beta.ppf(0.025, b+1, a-b+1)/p
ww = stats.beta.ppf(0.975, b+1, a-b+1)/p
ax.plot_surface(n,p,zz, rstride=1, cstride=1, linewidth=0, antialiased=False)
ax.plot_surface(n,p,ww, rstride=1, cstride=1, linewidth=0, antialiased=False)
plt.title("figure 7")
plt.show()
```

```python
from mpl_toolkits.mplot3d import axes3d, Axes3D 
fig = plt.figure()
ax = Axes3D(fig)

n = np.arange(1000, 20000, 1000)
p = np.arange(0.001, 0.05, 0.001)
n, p = np.meshgrid(n, p)

a = n 
b = n * p
zz = stats.beta.ppf(0.025, b+1, a-b+1)/p
ww = stats.beta.ppf(0.975, b+1, a-b+1)/p
ax.plot_surface(n,p,zz, rstride=1, cstride=1, linewidth=0, antialiased=False)
ax.plot_surface(n,p,ww, rstride=1, cstride=1, linewidth=0, antialiased=False)
plt.title("figure 8")
plt.show()
```

由图7与图8得出以下结论：
1. 随着曝光数与点击数的增加，置信区间会变窄。
2. 实际ctr较大的广告所需要的资源较少。

## 结论

针对广告平台实际情况，分别对高点击率（5%），中等点击率（0.6%），低点击率（0.1%）的广告，计算CTR探索实验曝光分别为100，500，1000，10000时的置信度：

```python
import numpy as np
import scipy.stats as stats

nn = [100, 500, 1000, 10000, 100000, 1000000]
pp = [0.05, 0.003, 0.005, 0.001]

for (i,n) in enumerate(nn):
    for (j,p) in enumerate(pp):
        a = n 
        b = n * p
        zz = stats.beta.ppf(0.025, b+1, a-b+1)/p
        ww = stats.beta.ppf(0.975, b+1, a-b+1)/p
        print n,p,ww-zz
```

可见，当点击率为0.006时，1000个探索曝光后，点击率的浮动范围会达到80%左右，10000个探索曝光能确保点击率的上下浮动不超过25%.
