```python
X = stats.beta.rvs(4, 498, size=50000)
 
plt.hist(X, 500, normed=True)
plt.title("figure 3")
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 5))
x = np.linspace(stats.beta.ppf(0.0, 4, 498), stats.beta.ppf(0.999, 4, 498), 100)
y = np.linspace(0, 0.999, 1000)
rv = stats.beta(4, 498)
markers_on = [25, 975]

plt.title("figure 5")
ax[0].plot(x, rv.cdf(x), 'k-', lw=1.5)
ax[1].plot(y, rv.ppf(y), 'k-', lw=1.5)
ax[1].annotate('lower bound:' + str(rv.ppf(0.16)), xy=(0.16, rv.ppf(0.16)), xytext=(0.03, 0.01),arrowprops=dict(facecolor='blue', shrink=0.05))
ax[1].annotate('upper bound:' + str(rv.ppf(0.84)), xy=(0.84, rv.ppf(0.84)), xytext=(0.9, 0.025),arrowprops=dict(facecolor='blue', shrink=0.05))
ax[1].plot([0.16, 0.16,], [0, rv.ppf(0.16),], 'k--', linewidth=2.5)
ax[1].plot([0.84, 0.84,], [0, rv.ppf(0.84),], 'k--', linewidth=2.5)
ax[1].plot([0, 0.16,], [rv.ppf(0.16), rv.ppf(0.16),], 'k--', linewidth=2.5)
ax[1].plot([0, 0.84,], [rv.ppf(0.84), rv.ppf(0.84),], 'k--', linewidth=2.5)
plt.show()
```
